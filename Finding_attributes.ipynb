```python
# Import the pandas library
import pandas as pd

import json


# Open the JSON file
with open(r'/content/sample_data/outputfull.json', 'r') as f:
    # Load the data from the file
    data = json.load(f)
```

```python
keyslist=[]
for item in data:
    p_attributes = item['p_attributes']
    # p_attributes is a string representation of a dictionary, so we need to convert it to a dictionary object
    p_attributes = eval(p_attributes)  # use caution when using eval()
    # get all the keys in the dictionary
    keys = p_attributes.keys()
    keyslist.extend(keys)
  
```

```python
len(keyslist)
```

```python
Uniquekeys=list(set(keyslist))
```

```python
len(Uniquekeys)
```

```python
# create a Pandas Series from the list
s = pd.Series(keyslist)
# get the count of each value in the Series
counts = s.value_counts()
# convert the Series to a DataFrame
df = counts.reset_index()
# rename the columns
df.columns = ['value', 'count']
```

```python
df_sorted = df.sort_values('count',ascending=False)
```
